# Extracting Journey Day and Month

```python
# Ectracting day of week from Date_of_Journey columns
df["Journey_day"]=df["Date_of_Journey"].dt.day_name()
df["Journey_day"]
```

```python
# Ectracting Month from Date_of_Journey columns
df["Journey_Month"]=df["Date_of_Journey"].dt.month_name()
df["Journey_Month"]
```

```python
#Summarizing the number of flights by day of week
no_of_flights_by_day_of_week=df.groupby("Journey_day")[["Airline"]].count().reset_index()

no_of_flights_by_day_of_week.columns=["Day_of_week","Number_of_Flights"]
weekday_order=["Monday","Tuesday","Wednesday",
               "Thursday","Friday","Saturday",
               "Sunday"]

plt.figure(figsize=(8,4))
sns.barplot(data=no_of_flights_by_day_of_week,y="Number_of_Flights",x="Day_of_week",order=weekday_order,palette="tab10")
plt.title("Number Of Flights By Day Of Week",fontsize=15)
plt.xlabel("Day Of week",fontsize=12)
plt.ylabel("Number Of flights",fontsize=12)
plt.show()

#Summarizing the number of flights by month
no_of_flights_by_month=df.groupby("Journey_Month")[["Airline"]].count().reset_index()

no_of_flights_by_month.columns=["Monthly","Number_of_Flights"]
month_order=["March","April","May","June"]

plt.figure(figsize=(8,4))
sns.barplot(data=no_of_flights_by_month,y="Number_of_Flights",x="Monthly",order=month_order,palette="tab20")
plt.title("Number Of Flights By Month",fontsize=15)
plt.xlabel("Monthly",fontsize=12)
plt.ylabel("Number Of flights",fontsize=12)
plt.show()
```

## Insights of 1st chart

Weekend days (Saturday and Sunday) show moderate flight volumes, indicating steady leisure travel.

### Flight frequency is highest mid-week and dips slightly on Fridays.

## Insights of 2nd chart
May records the highest number of flights, followed closely by June.

# Flight duration vs price

```python

import pandas as pd

def convert_to_minutes(duration):
    hours = 0
    minutes = 0
    if "h" in duration:
        hours = int(duration.split("h")[0])
        duration = duration.split("h")[1]
    if "m" in duration:
        minutes = int(duration.split("m")[0])
        duration = duration.split("m")[1]
    return hours * 60 + minutes
df['Duration_mins'] = df["Duration"].apply(convert_to_minutes)
df["Duration_mins"]

plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x="Duration_mins",y="Price",hue="Total_Stops")
plt.title("Flight Duration vs Ticket Price",fontsize=20)
plt.xlabel("Duration (minutes)",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

There is a clear positive relationship between flight duration and ticket price — as the duration increases, the price tends to rise.

Non-stop flights (typically represented by a distinct color in the hue) often have higher prices for the same or even shorter durations, reflecting 
the convenience premium passengers pay for direct connections.

Flights with one or more stops show wider price variation — some are cheaper due to layover inconvenience, while others are costly due to peak routes 
or limited competition.

A few outliers may exist — short-duration flights with high prices, possibly due to last-minute bookings or premium airlines operating those routes.

### Overall Insight:Ticket prices increase with flight duration, and non-stop flights command a higher average fare, highlighting how travel time and stopovers influence airfare pricing strategies.

# impact of additional info

```python
z=df.groupby("Additional_Info")[["Price"]].mean().sort_values("Price",ascending=False).reset_index()
z
z_encoded=pd.get_dummies(z,drop_first=True)
z_encoded
plt.figure(figsize=(8,4))
sns.heatmap(z_encoded.corr(),annot=True)
plt.title("Impact Of Additional info",fontsize=20,y=1.05)
plt.show()
```

The heatmap correlations help reveal how these encoded categories relate to one another — higher correlation values among certain features suggest 
that specific flight characteristics often occur together (e.g., long layovers and higher fares, or missing amenities and lower fares).

### Overall Insight:
The type of “Additional Info” provided significantly influences ticket prices.

Flights with premium services (e.g., Business class) tend to have higher fares,

while budget options (e.g., No meal, No info) reflect lower-cost offerings.
